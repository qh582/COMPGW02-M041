```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

const = pd.read_csv("constant_bidding_budget.csv")
random = pd.read_csv("random_bidding_budget.csv")
gate = pd.read_csv("gate_bidding_budget.csv")
```

```python
const1 = const[const['bid_price']==27]
```

```python
random1 = random[random['Upper_bound']==27]
```

```python
fig = plt.figure()
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="white", color_codes=True)
n_groups = 7
index = np.arange(n_groups)
bar_width = 0.35
opacity = 0.4
rects1 = plt.bar(index,const1.cost.values, bar_width,
                 alpha=opacity,
                 color='b',
                 label='Constant')
rects2 = plt.bar(index + bar_width,random1.cost.values , bar_width,
                 alpha=opacity,
                 color='r',
                 label='Random')
plt.xlabel('Budget')
plt.ylabel('Cost')
plt.title('Cost of bidding function', fontsize=16)
plt.xticks(index + bar_width / 2, const1.budget)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
fig.savefig('const_ran.eps')
```

```python
gate=gate[gate['norm']==2.95]#compromised value between ctr and num of clicks
const=gate[gate['const']==195]
random=gate[gate['random']==95]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="white", color_codes=True)
```

```python
random.budget.values
```

```python
fig = plt.figure()
plt.title('Total clicks Performance', fontsize=16)
plt.xlabel('budget')
plt.ylabel('total clicks')
plt.plot(random.clicks.values,label='random')
plt.plot(const.clicks.values,label='constant')
plt.plot(gate.clicks.values,label='gate')

plt.grid(True)
plt.xticks(range(len(random.budget.values)), random.budget.values)
plt.legend(loc='lower right')
plt.show()
fig.savefig('total_clicks.eps')
```

```python
fig = plt.figure()
plt.title('eCPC Performance', fontsize=16)
plt.xlabel('budget')
plt.ylabel('eCPC')
plt.plot(random.cpc.values,label='random')
plt.plot(const.cpc.values,label='constant')
plt.plot(gate.cpc.values,label='gate')

plt.grid(True)
plt.xticks(range(len(random.budget.values)), random.budget.values)
plt.legend(loc='upper right')
plt.show()
fig.savefig('ecpc.eps')
```

```python
fig = plt.figure()
plt.title('Impression Performance', fontsize=16)
plt.xlabel('budget')
plt.ylabel('impression')
plt.plot(random.cpc.values,label='random')
plt.plot(const.cpc.values,label='constant')
plt.plot(gate.cpc.values,label='gate')

plt.grid(True)
plt.xticks(range(len(random.budget.values)), random.budget.values)
plt.legend(loc='upper right')
plt.show()
fig.savefig('impression.eps')
```
